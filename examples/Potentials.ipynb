```python
import numpy as np
import pylab as plt
import potentials
import constants
```

```python
R=np.linspace(1e-5,0.005,100)
plt.figure(figsize=(5,5),dpi=150)
plt.plot(R*constants.hcInEVAngstrom,potentials.Silvera_Triplet(R),label='Silvera Triplet')
plt.plot(R*constants.hcInEVAngstrom,potentials.Silvera_Singlet(R),label='Silvera Singlet')
plt.legend(loc='upper right')
plt.ylim(-10,10)
plt.xlabel(r"R ($\AA$)")
plt.ylabel("V (eV)")
```

```python
Pots=[potentials.Silvera_Triplet,potentials.Silvera_Triplet2, potentials.UnCorrectedSilvera_Triplet]
Names=['Silvera 1','Silvera 2','Un-Adiabatic Corrected Silvera']
R=np.linspace(0,0.01,100)
plt.figure(figsize=(5,5),dpi=150)
for i in range(0,len(Pots)):
    plt.plot(R*constants.hcInEVAngstrom, Pots[i](R)/Pots[0](R),label=Names[i])
plt.ylim(0.7,1.1)
plt.xlabel(r"R ($\AA$)")
plt.ylabel("Ratio")
plt.legend(loc='upper right')
```
