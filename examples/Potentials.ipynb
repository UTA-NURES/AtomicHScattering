```python

import numpy as np
import pylab as plt
import potentials
import constants
```

```python
R=np.linspace(1e-5,0.005,100)
plt.figure(figsize=(5,5),dpi=150)
plt.plot(R*constants.hcInEVAngstrom,potentials.Silvera_Triplet(R),label='Silvera Triplet')
plt.plot(R*constants.hcInEVAngstrom,potentials.Silvera_Singlet(R),label='Silvera Singlet')
plt.legend(loc='upper right')
plt.ylim(-10,10)
plt.xlabel(r"R ($\AA$)")
plt.ylabel("V (eV)")
```

```python
Pots=[potentials.Silvera_Triplet,potentials.Silvera_Triplet2, potentials.UnCorrectedSilvera_Triplet, potentials.Jamieson_TripletCombo]
Names=['Silvera Triplet 1','Silvera Triplet 2','Un-Adiabatic Corrected Silvera 1','Jamieson Triplet']
R=np.linspace(0,0.0025,100)
plt.figure(figsize=(5,5),dpi=150)
for i in range(0,len(Pots)):
    plt.plot(R*constants.hcInEVAngstrom, Pots[i](R),label=Names[i])
plt.ylim(-1,6)
plt.xlim(0,5)
plt.xlabel(r"R ($\AA$)")
plt.ylabel("Ratio")
plt.legend(loc='upper right')
```

```python
Pots=[potentials.Silvera_Triplet,potentials.Silvera_Triplet2, potentials.UnCorrectedSilvera_Triplet, potentials.Jamieson_TripletCombo]
Names=['Silvera 1','Silvera 2','Un-Adiabatic Corrected Silvera','Jamieson']
R=np.linspace(0,0.01,100)
plt.figure(figsize=(5,5),dpi=150)
for i in range(0,len(Pots)):
    plt.plot(R*constants.hcInEVAngstrom, Pots[i](R)/Pots[0](R),label=Names[i])
plt.ylim(0.7,1.1)
plt.xlabel(r"R ($\AA$)")
plt.ylabel("Ratio")
plt.legend(loc='lower right')
```

```python
Pots = [ potentials.Kolos_SingletCombo,potentials.Silvera_Singlet, potentials.Jamieson_SingletCombo]
Names = ['Kolos Singlet','Silvera Singlet','Jamieson Singlet']
linetype=['-',':','--','-.','-']
R = np.linspace(0, 0.0025, 100)+1e-5
plt.figure(figsize=(5, 5), dpi=150)
for i in range(0, len(Pots)):
    plt.plot(R * constants.hcInEVAngstrom, Pots[i](R),linetype[i], label=Names[i],alpha=0.5)
plt.xlim(0,5)
plt.ylim(-6,6)
plt.xlabel(r"R ($\AA$)")
plt.ylabel("Ratio")
plt.legend(loc='upper right')
```

```python
R = np.linspace(0, 0.02, 100)+1e-5


plt.plot(R*constants.hcInEVAngstrom,np.abs(potentials.Silvera_Triplet(R)),label='Silvera Triplet')

plt.plot(R*constants.hcInEVAngstrom,potentials.DipoleRadialPart(R),label='Dipole Radial Part')
plt.semilogy()
plt.legend(loc='upper right')
plt.ylabel("Potential (eV)")
plt.xlabel(r"Distance ($\AA$)")
```
