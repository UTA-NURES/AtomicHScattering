```python

import numpy as np
import pylab as plt
import matplotlib as mpl
import potentials
import constants
```

```python


R=np.linspace(1e-5,0.005,100)
plt.figure(figsize=(5,5),dpi=150)
plt.plot(R*constants.hcInEVAngstrom,potentials.Silvera_Triplet(R),label='Silvera Triplet')
plt.plot(R*constants.hcInEVAngstrom,potentials.Silvera_Singlet(R),label='Silvera Singlet')
plt.legend(loc='upper right')
plt.ylim(-10,10)
plt.xlabel(r"R ($\AA$)")
plt.ylabel("V (eV)")
```

```python
R=np.linspace(0,0.0025,100)

Singlets = potentials.Singlets.keys()
linetype=['-',':','--','-.','-']
R = np.linspace(0, 0.0025, 100)+1e-5
plt.figure(figsize=(4, 5), dpi=150)
colors=['DarkRed','red','DarkBlue','blue','DarkGreen']

colors=['DarkRed','red','dimgrey','orange','DarkGreen']
Triplets=potentials.Triplets.keys()

for i in Triplets:
    plt.plot(R*constants.hcInEVAngstrom, potentials.Triplets[i](R),label=i)

for i in Singlets:
    plt.plot(R * constants.hcInEVAngstrom, potentials.Singlets[i](R),'--', label=i)
plt.xlim(0,4)
plt.ylim(-6,6)
plt.xlabel(r"R ($\AA$)")
plt.ylabel("Potential (V)")
plt.legend(loc='upper right',ncol=2,fontsize=8,title='     Triplets             Singlets')

plt.show()
```

```python

plt.figure(figsize=(5, 5), dpi=150)
R = np.linspace(0.00001, 0.02,1000)


plt.plot(R*constants.hcInEVAngstrom,np.abs(potentials.Silvera_Triplet(R)),label='Central Triplet',color='DarkBlue')

plt.plot(R*constants.hcInEVAngstrom,potentials.DipoleRadialPart(R),'--',label='Dipole Potential',color='DarkRed')
plt.plot(R*constants.hcInEVAngstrom,np.abs((2/3*potentials.CorrInterp(R))*potentials.Silvera_Triplet(R)),':',color='DarkGreen',label='Adiabatic Correction H->T')

plt.semilogy()
plt.legend(loc='upper right')
plt.ylabel("|V| (eV)")
plt.xlim(0,20)
plt.ylim(1e-9,100)
plt.xlabel(r"Distance ($\AA$)")
```
