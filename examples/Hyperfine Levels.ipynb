```python
# Example script to calculate and plot the hyperfine / Zeeman levels of T and H
```

```python
import sys
sys.path.append("../src/")
sys.path.append("../src/AtomicH")
```

```python
from AtomicH import hyperfine, constants
import numpy as np
import pylab as plt
```

```python
B_values = np.logspace(-3,1,50)
HLevels = hyperfine.AllHFLevels(B_values, constants.HydrogenConstants)
TLevels = hyperfine.AllHFLevels(B_values, constants.TritiumConstants)
```

```python
plt.figure(figsize=(5,5),dpi=150)
cols={'a':'DarkRed','b':'DarkBlue','c':'DarkGreen','d':'dimgrey'}
for l in ['a','b','c','d']:
    plt.plot(B_values, HLevels[l],label='H'+str(l),color=cols[l],alpha=0.5)
    plt.plot(B_values, TLevels[l],'--',label='T'+str(l),color=cols[l],alpha=1)
plt.xlim(0,0.1)
plt.ylim(-1e-5,1e-5)
plt.title(r"Zeeman-Hyperfine levels")
plt.legend(loc='upper left',ncol=4)
plt.ylabel("U (eV)")
plt.xlabel("B (T)")
```
