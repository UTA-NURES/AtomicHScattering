```python
import pandas as pd
import pylab as plt
```

```python
DipoleChannels=[]
DipoleChannels.append({'alpha':'d','beta':'d','alphaprime':'a','betaprime':'a'})
DipoleChannels.append({'alpha':'d','beta':'d','alphaprime':'a','betaprime':'c'})
DipoleChannels.append({'alpha':'d','beta':'d','alphaprime':'a','betaprime':'d'})
DipoleChannels.append({'alpha':'d','beta':'d','alphaprime':'c','betaprime':'c'})
DipoleChannels.append({'alpha':'d','beta':'d','alphaprime':'c','betaprime':'d'})
```

```python
dat=pd.read_csv("./Tables/SilveraDipoles.csv")
```

```python
datH=dat[dat.Isotope=="H"]
datT=dat[dat.Isotope=="T"]
```

```python
for c in DipoleChannels:
    plt.figure(figsize=(5,5),dpi=250)

    datCH=datH[datH.Channel==str(c)]
    for T in datCH[('T')].unique():
        datCHT=datCH[datCH[('T')]==T]
        plt.plot(datCHT.B.values,datCHT.sigma.values,label=T)
    plt.legend(ncol=3,title="Temperature (K)")
    Title=r"Hydrogen, "+ c['alpha']+c['beta']+r"$\rightarrow$"+c['alphaprime']+c['betaprime']
    plt.title(Title)
    plt.loglog()
    plt.ylabel(r"$\sigma$ (a.u.)$^2$")
    plt.xlabel(r"B(T)")

    plt.ylim(1e-8,1)
    plt.xlim(1e-3,10)
    plt.grid(which='both')
    fname=r"Hydrogen_"+ c['alpha']+c['beta']+r"_to_"+c['alphaprime']+c['betaprime']+".png"
    plt.savefig("Plots/"+fname,dpi=250,bbox_inches='tight')
    plt.show()







```

```python
for c in DipoleChannels:
    plt.figure(figsize=(5,5),dpi=250)


    datCT=datT[datT.Channel==str(c)]
    for T in datCT[('T')].unique():
        datCTT=datCT[datCT[('T')]==T]
        plt.plot(datCTT.B.values,datCTT.sigma.values,'-',label=T)
    plt.legend(ncol=3,title="Temperature (K)")
    Title=r"Tritium, "+ c['alpha']+c['beta']+r"$\rightarrow$"+c['alphaprime']+c['betaprime']
    plt.title(Title)
    plt.ylabel(r"$\sigma$ (a.u.)$^2$")
    plt.xlabel(r"B(T)")
    plt.loglog()
    plt.ylim(1e-6,100)
    plt.xlim(1e-3,10)
    plt.grid(which='both')
    fname=r"Tritium_"+ c['alpha']+c['beta']+r"_to_"+c['alphaprime']+c['betaprime']+".png"
    plt.savefig("Plots/"+fname,dpi=250,bbox_inches='tight')
    plt.show()
```
